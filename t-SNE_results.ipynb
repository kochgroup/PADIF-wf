### t-SNE Results

```python
import os
import pandas as pd
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
target_name = "ADA"
os.chdir(f"/nfs/home/dvictori/Documents/Dockings_results/mols_prep/{target_name}")
target = pd.read_csv(f"{target_name}_PADIF-1.csv", sep=",", index_col=0)
target
```

```python
tar_val = target.drop(columns=["Activity"])
tar_val = tar_val.values.tolist()

tSNE = TSNE(n_components=2, perplexity=50, n_iter=2000, learning_rate=300, random_state=1, metric ="jaccard").fit_transform(tar_val)
target['x'] = tSNE[:, 0]
target['y'] = tSNE[:, 1]

kws = dict(s=50, linewidth=.5, edgecolor="w")
fg = sns.FacetGrid(data=target, hue='Activity', aspect=1.61, size=7, palette=['lime','dodgerblue'])
fg.map(plt.scatter, 'x', 'y', **kws).add_legend()
```
